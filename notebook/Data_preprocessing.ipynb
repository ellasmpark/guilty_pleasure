# Recipe Data Preprocessing : Recipe data to recommend to the user


```python
#Opening raw_recipes csv file in the raw data folder

BASE = os.path.dirname(os.path.abspath('../RAW_recipes.csv'))
file_path = os.path.join(BASE, 'raw_data', 'RAW_recipes.csv')

df=pd.read_csv(file_path)


df.head(3)
```

```python
new_df=df.drop(columns=['name','id','minutes','contributor_id','submitted','tags','n_steps','description'])
new_df.head(3)
```

```python
# Breaking down nutrition column into individual macros columns
import json
```

```python
calories=[]
fats=[]
protein=[]
carbs=[]
for i in new_df['nutrition']:
    i=json.loads(i)  #converting string representation of a list to an actual list
    calories.append(i[0])
    fats.append(i[1])
    protein.append(i[4])
    carbs.append(i[6])
```

```python
new_df['calories']=calories
new_df['fats']=fats
new_df['protein']=protein
new_df['carbs']=carbs
new_df=new_df.drop(columns=['nutrition','steps','n_ingredients'])
new_df
```

```python
import ast
```

```python
# Convert string representations of lists to actual lists using map() and ast.literal_eval()
new_df['ingredients'] = new_df['ingredients'].map(ast.literal_eval)

# Verify the updated data type
print(new_df['ingredients'].dtype)  # Output: object (if the dtype is still showing as object)

# Convert the column to a list data type
new_df['ingredients'] = new_df['ingredients'].map(list)

# Verify the updated data type
print(new_df['ingredients'].dtype)  # Output: object (if the dtype is still showing as object)

# ingredients=[]
# for j in new_df['ingredients']:
#     j= ast.literal_eval(j)  #converting string representation of a list to an actual list
#     ingredients.append(j)
# new_df['ingredients']=ingredients
```

```python
new_df.to_csv('preprocessed.csv',  index=False, header=True)
```
